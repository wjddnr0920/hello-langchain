# 1. 웹 문서
### WebBaseLoader 이용하여 웹 페이지 데이터 가져오기


```python
import bs4
from langchain_community.document_loaders import WebBaseLoader

# 여러 개의 url 지정 가능
url1 = "https://blog.langchain.dev/customers-replit/"
url2 = "https://blog.langchain.dev/langgraph-v0-2/"

loader = WebBaseLoader(
    web_paths=(url1, url2),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("article-header", "article-content")
        )
    ),
)

docs = loader.load()
len(docs)
```

```python
docs[0]
```

# 3. 디렉터리 폴더
### DirectoryLoader 이용하여 특정 폴더의 모든 파일을 가져오기

```python
# Python에서는 glob 사용
import os
from glob import glob

files = glob(os.path.join(r'E:\Portfolio\Study\hello-langchain\rag_data', '*.txt'))
files
```

```python
# LangChain에서는 DirectoryLoader 사용
from langchain_community.document_loaders import DirectoryLoader, TextLoader

loader = DirectoryLoader(path=r'E:\Portfolio\Study\hello-langchain\rag_data', glob="*.txt", loader_cls=TextLoader, loader_kwargs={"encoding": "utf-8"})

data = loader.load()

len(data)
```

# 4. CSV 문서

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(file_path=r"E:\Portfolio\Study\hello-langchain\rag_data\한국주택금융공사_주택금융관련_지수_20160101.csv", encoding='cp949')
data = loader.load()

len(data)
```

```python
# 데이터 출처 정보를 특정 필드(열, column)로 지정, metadata의 source가 바뀜
loader = CSVLoader(file_path=r"E:\Portfolio\Study\hello-langchain\rag_data\한국주택금융공사_주택금융관련_지수_20160101.csv", encoding='cp949',
                   source_column='연도')

data = loader.load()

data[0]
```

```python
# CSV 파싱 옵션을 지정
loader = CSVLoader(file_path=r"E:\Portfolio\Study\hello-langchain\rag_data\한국주택금융공사_주택금융관련_지수_20160101.csv", encoding='cp949',
                   csv_args={
                       'delimiter': '\n',
                   })

data = loader.load()

data[0]
```
