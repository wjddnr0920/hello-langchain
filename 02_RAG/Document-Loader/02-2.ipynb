## 형식이 없는 PDF 문서 로드
### UnstructuredPDFLoader 이용하여 PDF 파일 데이터 가져오기
- PDF 파일 내의 다양한 텍스트 조각(chunk)를 서로 다른 "elements"로 생성하고, 
- 별도 설정을 하지 않으면 기본적으로 이러한 elements를 결합하여 하나의 텍스트 데이터로 반환합니다.

**별도 프로그램을 설치해야돼서 포기**

```python
from langchain_community.document_loaders import UnstructuredPDFLoader

pdf_filepath = r'E:\Portfolio\Study\hello-langchain\rag_data\000660_SK_2023.pdf'

# 전체 텍스트를 단일 문서 객체로 변환
loader = UnstructuredPDFLoader(pdf_filepath)
pages = loader.load()

len(pages)  # 1
```

## PDF 문서의 메타 데이터를 상세하게 추출
### PyMuPDFLoader 이용하여 PDF 파일 데이터 가져오기

```python
from langchain_community.document_loaders import PyMuPDFLoader

pdf_filepath = r'E:\Portfolio\Study\hello-langchain\rag_data\000660_SK_2023.pdf'

loader = PyMuPDFLoader(pdf_filepath)
pages = loader.load()

len(pages)
```

## 온라인(on-line) PDF 문서 로드
### OnlinePDFLoader 이용하여 온라인 PDF 파일의 데이터를 가져오기

```python
from langchain_community.document_loaders import OnlinePDFLoader

# Transformers 논문을 로드
loader = OnlinePDFLoader("https://arxiv.org/pdf/1706.03762.pdf")
pages = loader.load()

len(pages)
```

## 특정 폴더의 모든 PDF 문서 로드
### PyPDFDirectoryLoader 이용하여 특정 폴더의 모든 PDF 파일을 가져오기

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

loader = PyPDFDirectoryLoader(r'E:\Portfolio\Study\hello-langchain\rag_data')
data = loader.load()

len(data)  # 100 <- 21page + 79page
```
