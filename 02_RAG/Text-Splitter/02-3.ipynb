# RecursiveCharacterTextSplitter
- 텍스트를 재귀적으로 분할하여 의미적으로 관련 있는 텍스트 조각들이 같이 있도록 하는 목적으로 설계

- 문자 리스트(`['\n\n', '\n', ' ', '']`)의 문자를 순서대로 사용하여 텍스트를 분할

- 분할된 청크들이 설정된 `chunk_size`보다 작아질 때까지 이 과정을 반복

**파라미터**
- chunk_size: 각 청크의 최대 길이

- chunk_overlap: 인접한 청크 사이에 중복으로 포함될 문자의 수

- length_function: 청크의 길이를 계산하는 함수. 여기서는 len 함수가 사용되었으므로, 문자열의 길이를 기반으로 청크의 길이를 계산

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader(r'E:\Portfolio\Study\hello-langchain\rag_data\history.txt', encoding='utf-8')
data = loader.load()

print(len(data[0].page_content))
data[0].page_content
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    length_function=len
)

texts = text_splitter.split_text(data[0].page_content)

len(texts)
```

```python
len(texts[0]), len(texts[1]), len(texts[2])
```

**문장이 온전하게 유지된 채로 나누어지는 것을 확인할 수 있습니다.**

# 토큰 수를 기준으로 텍스트 분할 (Tokenizer 활용)
- LLM 모델에 적용되는 토크나이저를 기준으로 텍스트를 토큰으로 분할

- 이 토큰들의 수를 기준으로 텍스트를 청크로 나누면 모델 입력 토큰 수 조절 가능

**비교**

- `CharacterTextSplitter`를 사용하는 경우: 텍스트는 Splitter에 의해서만 분할되고 tiktoken 토크나이저는 분할된 텍스트를 병합하는 데 사용

- `RecursiveCharacterTextSplitter`를 사용하는 경우: 분할된 텍스트가 언어 모델에서 허용하는 토큰의 청크 크기보다 큰 경우 재귀적으로 분할

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=500,
    chunk_overlap=100,
    encoding_name="cl100k_base"
)

docs = text_splitter.split_documents(data)
len(docs)
```

```python
len(docs[0].page_content), len(docs[1].page_content), len(docs[2].page_content)
```

```python
docs[0]
```

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    model_name="gpt-4",
    chunk_size=500,
    chunk_overlap=100,
)

docs = text_splitter.split_documents(data)
len(docs)
```
